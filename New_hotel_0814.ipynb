```python
import folium
import pandas as pd
from folium import IFrame

#data = cdata.iloc[0:200,:]   #0到200行，所有列
data = pd.read_csv('new_hotel_0813.csv', header = 0, encoding="utf_8_sig")
lat_cn = 35
lng_cn = 103

data_low = data.loc[data["Tier_Brand"]=="经济"]
data_mh = data.loc[data["Tier_Brand"]=="中高端"]


incidents = folium.map.FeatureGroup()

for lat, lng, in zip(data_low.lat, data_low.lng):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=12, # define how big you want the circle markers to be
            color='white',
            fill=True,
            fill_color='lightblue',
            fill_opacity=0.5
        )
    )

for lat, lng, in zip(data_mh.lat, data_mh.lng):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=12, # define how big you want the circle markers to be
            color='white',
            fill=True,
            fill_color='red',
            fill_opacity=0.5
        )
    )


# Add incidents to map
bj_map = folium.Map(location=[lat_cn, lng_cn], zoom_start=3.6)
bj_map.add_child(incidents)


latitudes = list(data.lat)
longitudes = list(data.lng)
labels = []
def utf2asc(s):
    return str(str(s).encode('ascii', 'xmlcharrefreplace'))[2:-1]

for i in list(data.name):
    heading3 = """<h5>{}</h5>""".format
    iframe = IFrame(html=heading3(utf2asc(i)),width=180,height=50)
    popup = folium.Popup(iframe)
    labels.append(popup)
#labels = list(data.name)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label, color = 'red').add_to(bj_map)    
    
# add incidents to map
bj_map.add_child(incidents)
```
